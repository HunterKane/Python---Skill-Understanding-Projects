```python
# Import libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Keep within plot
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Programming
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title');
```

```python
# Create multiple plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b');
```

```python
# Object Orientated Methods
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y_label')
axes.set_title('Title');
```

```python
# Multiple plots 

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_xlabel("X Label")
axes1.set_ylabel("Y Label")
axes1.set_title("Larger Plot")


axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.set_xlabel("X Label")
axes2.set_ylabel("Y Label")
axes2.set_title("Smaller Plot")

axes1.plot(x,y)
axes2.plot(y,x);
```

```python
# Create subplots 

fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

## Figure Size, and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x);

plt.tight_layout()
```

```python
# Create figure with legend / custom line colors/line width/line style/markers/markers size/marker colors 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y ** 2, color = 'purple',lw=3,ls= '--', marker= 'o',label = 'X Squared')
ax.plot(x,x ** 3, color = 'green', lw=1, marker='*',markersize=10,
        markerfacecolor='red', label = 'C Cubed')
ax.set_title('Title', fontsize = 25)

#set legend with best location 

ax.legend(loc=0); # also create custom location = loc(0.1,0.1)
```

### Exercise 1: Read Total profit of all months and show it using a line plot:

- X label name = Month Number
- Y label name = Total profit

```python
# Load in data 
df = pd.read_csv('salesData.txt')
```

```python
df.head()
```

```python
df['total_profit'].plot(figsize = (15,8))
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Profit Per month", fontsize = 20);
```

### Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

```python
profitList = df['total_profit'].tolist()
monthList = df['month_number'].tolist()

plt.plot(monthList, profitList, label = "Profit Data for The Past Year",
        color = 'red', marker='o', markerfacecolor='k',
        ls='--', lw=3)

# add labels

plt.x_label=('Month Number')
plt.ylabel('Profit in Dollars')
plt.legend(loc='lower right')

plt.title('Comapny Sales Data For Previous Year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### Exercise Question 3: Read all product sales data and show it  using a multiline plot

- Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
# Data needed for plot / tolist 
monthList = df['month_number'].tolist()
faceCreamSale = df['facecream'].tolist()
faceWashSale = df['facewash'].tolist()
toothPasteSale = df['toothpaste'].tolist()
bathingSoapSale = df['bathingsoap'].tolist()
shampooSale = df['shampoo'].tolist()
moisturizerSale = df['moisturizer'].tolist()

#plot
plt.plot(monthList, faceCreamSale, label = "Monthly Facecream Sale")
plt.plot(monthList, faceWashSale, label = "Monthly facecream Sales", marker = 'o')
plt.plot(monthList, toothPasteSale, label = "Monthly toothpaste Sales", marker = 'o')
plt.plot(monthList, bathingSoapSale, label = "Monthly bathingsoap Sales", marker = 'o')
plt.plot(monthList, shampooSale, label = "Monthly Shampoo Sale", marker = 'o')
plt.plot(monthList, moisturizerSale, label = "Monthly Moisturizer Sale", marker = 'o')


plt.xlabel("Monthly Numbers")
plt.ylabel("Sales units in numbers")
plt.legend(loc=0)
plt.xticks(monthList)
plt.title('Sales Data')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000])
plt.show()
```

### Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should .

```python
# Data needed for plot / tolist 
monthList = df['month_number'].tolist()
toothPasteSale = df['toothpaste'].tolist()

# plot 
plt.scatter(monthList, toothPasteSale,label = 'toothpaste Sales')

plt.xlabel("Monthly Numbers")
plt.ylabel("Toothpaste Sales units in numbers")
plt.xticks(monthList)
plt.legend(loc = 0)
plt.title('Toothpaste Sales Data')
plt.grid(True, lw=1, ls = '--')
```

### Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
# Data needed / to list 
monthList = df['month_number'].tolist()
faceCreamSale = df['facecream'].tolist()
faceWashSale = df['facewash'].tolist()

plt.bar([a-0.25 for a in monthList],faceCreamSale, width = 0.50, label = 'Face Cream ')
plt.bar([a+0.25 for a in monthList], faceWashSale, width = 0.50, label = 'Face Wash', color = 'r')
plt.yticks([1000, 2000, 3000, 4000, 5000])
plt.legend(loc = 0);

plt.xlabel('Monthly Number')
plt.ylabel('Sales units in number')
plt.title('Sales Data')
plt.grid(True, lw= 1, ls ='--')
```
