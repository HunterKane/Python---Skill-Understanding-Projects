```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Different types of visualizations 

- Distribution 
- Categorical 
- Matrix
- Grids
- Regression Plots

### Distribution 

```python
# Load in built in Seaborn dataset 
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins=30);
```

```python
sns.kdeplot(tips['total_bill']);
```

- Most of the bills are between 10-20 dollars. Fades away the higher the bill price. 

```python
sns.jointplot(x='total_bill', y='tip', data = tips, kind = 'scatter')
```

- Higher bill normally means a higher tip due to the bill price is higher. Data shows an upward trend. 

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.rugplot(tips['total_bill']);
```

### Catigorical 

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std);
```

```python
sns.countplot(x='sex', data=tips);
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker');
```

- People on fridays tends to have a larger bill who are not smokers.
- Poeple on Saturday and Sunday are smokers with a larger bill. 

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split = True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips,jitter = True, hue='sex', dodge=True);
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips, color='black')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind = 'bar');
```
